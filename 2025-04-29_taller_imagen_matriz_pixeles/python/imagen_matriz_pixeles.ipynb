```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('jirafa.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

## Acceder y Mostrar los Canales RGB y HSV

```python
# Separar canales RGB
r, g, b = cv2.split(img)

# Convertir a HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(img_hsv)

# Mostrar cada canal
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(r, cmap='Reds')
plt.title('Canal Rojo')

plt.subplot(2, 2, 2)
plt.imshow(g, cmap='Greens')
plt.title('Canal Verde')

plt.subplot(2, 2, 3)
plt.imshow(b, cmap='Blues')
plt.title('Canal Azul')

plt.subplot(2, 2, 4)
plt.imshow(h, cmap='hsv')
plt.title('Canal Hue')

plt.show()
```

## Modificar Regiones Espec√≠ficas de la Imagen

```python
# Cambiar el color de un √°rea rectangular:
# Definir las coordenadas de la regi√≥n
x1, y1 = 100, 100
x2, y2 = 200, 200

region = img[y1:y2, x1:x2].copy()

# Cambiar color a
img[y1:y2, x1:x2] = [0, 255, 183]  # amarillo

# Sustituir una regi√≥n por otra parte de la imagen:
# Copiar una regi√≥n de la imagen y pegarla en otra posici√≥n

img[y1+200:y2+200, x1-300:x2-300] = region

plt.imshow(img)
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calcular y visualizar el histograma de intensidades con `cv2.calcHist()`
hist = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
plt.plot(hist)
plt.title('Histograma de Intensidad')
plt.show()
```

## Aplicar ajustes de brillo y contraste de forma manual

```python
# Ajuste manual: Brillo y Contraste (Ecuaci√≥n)
alpha = 1.2  # Contraste
beta = 20    # Brillo
img_adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

# Mostrar la imagen ajustada
img_adjusted_rgb = cv2.cvtColor(img_adjusted, cv2.COLOR_BGR2RGB)
plt.imshow(img_adjusted_rgb)
plt.show()
```
